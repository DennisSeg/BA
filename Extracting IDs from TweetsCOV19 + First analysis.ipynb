```python
# Extract Tweet IDs from TweetsCOV19 into CSV
import csv,sys,json,carmen

#Path for TweetsCOV19
pathTweetsCOV19='C:/Users/dennis/Desktop/BachelorArbeit/TweetsCOV19.tsv'
#Path for new file with Tweet IDs
pathIDs='C:/Users/dennis/Desktop/tweetIDs.csv'


tweetIDs=[]
with open(pathTweetsCOV19, 'r', newline='',encoding="utf-8") as file:
    csv.field_size_limit(100000000)
    reader = csv.reader(file, delimiter='\t')
    for line in reader:
        tweetID=line[0]
        tweetIDs.append(tweetID)
        

with open (pathIDs, 'w', newline='',encoding="utf-8") as csvfile:   
        writer = csv.writer(csvfile, delimiter=';')
        for tweetID in tweetIDs:
            writer.writerow([tweetID])
            
# CSV file can be directly uploaded to application Hydrator
```

```python
# Count of Tweets with features "Geo", "Place", "User Location"

import csv
import json

# Path to hydrated TweetsCOV19 file
fileHydrated = 'C:/Users/dennis/Desktop/Bachelorarbeit/HydratedALL.jsonl'

count_coordinates=0
count_location=0
count_place=0
count=0


with open(fileHydrated,encoding="utf8",errors='ignore') as hydratedJSON:
    for line in hydratedJSON:
        
        line=json.loads(line)
        print(line)
        break
        if line["coordinates"] is not None:            
            count_coordinates=count_coordinates+1
            
        if line["user"]["location"]:
            count_location=count_location+1
            
        if line["place"] is not None:
            count_place=count_place+1        
        count=count+1
print(count)        
print("Geo-coordinates: "+str(count_coordinates))
print("Place: "+str(count_location))
print("User location: "+str(count_place))
```

```python
# Plots to compare monthly tweets count between TweetsCOV19 and hydrated TweetsCOV19 dataset

import csv
import json
import pandas as pd

# Path to both dataset
tweetsCOV='C:/Users/Dennis/Desktop/BachelorArbeit/TweetsCOV19.tsv'
hydratedtweetsCOV='C:/Users/dennis/Desktop/BachelorArbeit/HydratedALL.jsonl'

df_tweetCOV = pd.read_csv(tweetsCOV, delimiter ='\t',names=['id','username',"timestamp","Follower","Friends","Retweets","Favorites","Entities","Sentiment","Mentions","Hashtags","URLs"])

hydratedTimestamps=[]
with open(hydratedtweetsCOV,encoding="utf8",errors='ignore') as hydratedJSON:
    for line in hydratedJSON:
        line=json.loads(line)
        hydratedTimestamps.append(line["created_at"])
        
# Create Timestamp dictionary for TweetsCOV19 dataset
import datetime
import matplotlib.pyplot as plot

df_timestamp=df_tweetCOV["timestamp"]

month_countTweetsCOV={}

for x in df_timestamp:
    time = datetime.datetime.strptime(x,'%a %b %d %H:%M:%S +%f %Y')
    month=time.strftime("%B")
    if month =="September":
        continue
    if month not in month_countTweetsCOV:
        month_countTweetsCOV[month]=1
    else:
        month_countTweetsCOV[month]=month_countTweetsCOV[month]+1

# Create Timestamp dictionary for Hydrated TweetsCOV19 dataset        
month_countHydrated={}

for timestamp in hydratedTimestamps:
    time = datetime.datetime.strptime(timestamp,'%a %b %d %H:%M:%S +%f %Y')
    month=time.strftime("%B")
    if month =="September":
        continue
    if month not in month_countHydrated:
        month_countHydrated[month]=1
    else:
        month_countHydrated[month]=month_countHydrated[month]+1
        
# Create Plots for both timestamp dictionaries        
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

listTweetCov=list(month_countTweetsCOV.values())
listHydrated=list(month_countHydrated.values())
listMonths=list(month_countTweetsCOV.keys())


x = np.arange(len(listMonths))  # the label locations

width = 0.3  # the width of the bars

fig, ax = plt.subplots()

ax.bar(x - width/2, listTweetCov, width, label='TweetsCOV19')
ax.bar(x + width/2, listHydrated, width, label='Hydrated')
#ax.set_ylabel('Count')
ax.set_title('Monthly Tweet Distribution')
ax.set_xticks(x)
ax.set_xticklabels(listMonths, fontsize=8)
ax.legend()

fig.tight_layout()

plt.show()
```
